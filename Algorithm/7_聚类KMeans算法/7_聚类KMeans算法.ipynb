# 使用KMeans模型数据探索聚类

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets  import make_blobs
from sklearn.metrics import calinski_harabasz_score

# 构建数据集
x, y = make_blobs(n_samples = 100, n_features=2, centers=[[-1, -1],[0, 0],[1, 1],[2, 2]],cluster_std=[0.4, 0.2, 0.3, 0.2])
plt.figure()
plt.scatter(x[:, 0], x[:, 1])
plt.show()

# 模训练预测
model = KMeans(n_clusters=4)
y_prd = model.fit_predict(x)
plt.figure()
plt.scatter(x[:, 0], x[:, 1], c = y_prd)
plt.show()

# 评估
print(calinski_harabasz_score(x, y_prd))
```

# 聚类效果评估——误差平方和SSE

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score

# 构建数据集
x, y = make_blobs(n_samples = 100, n_features=2, centers=[[-1, -1],[0, 0],[1, 1],[2, 2]],cluster_std=[0.4, 0.2, 0.3, 0.2])

# 迭代不同k值，获取sse
temp_list = []
for k in range(1, 100):
    model = KMeans(n_clusters=k)
    model.fit(x)
    temp_list.append(model.inertia_)

# 绘制图像
plt.figure()
plt.grid()
plt.plot(range(1, 100),temp_list,'or-')
plt.show()
```

# 聚类效果评估——SC系数

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

x, y = make_blobs(n_samples = 100, n_features=2, centers=[[-1, -1],[0, 0],[1, 1],[2, 2]],cluster_std=[0.4, 0.2, 0.3, 0.2])

temp_list = []
for k in range(2, 100):
    model = KMeans(n_clusters = k, n_init = 'auto')
    model.fit(x)
    y_pre = model.predict(x)
    temp_list.append(silhouette_score(x, y_pre))

plt.figure()
plt.grid()
plt.plot(range(2, 100), temp_list, 'or-')
plt.show()
```

# 聚类效果评估——CH系数

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.metrics import calinski_harabasz_score
from sklearn.cluster import KMeans

x, y = make_blobs(n_samples = 100, n_features=2, centers=[[-1, -1],[0, 0],[1, 1],[2, 2]],cluster_std=[0.4, 0.2, 0.3, 0.2])

temp_list = []
for k in range(2, 100):
    model = KMeans(n_clusters=k, n_init='auto')
    model.fit(x)
    y_pre = model.predict(x)
    temp_list.append(calinski_harabasz_score(x, y_pre))

plt.figure()
plt.grid()
plt.plot(range(2, 100), temp_list, 'or-')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import pandas as pd
"""性别：热编码
年龄，标准化"""

# 读取数据
data = pd.read_csv('./data/customers.csv')

# 特征选择
x = data.iloc[:, [3, 4]]

# # 模型训练
# # K值的选择
# temp_list = []
# for i in range(1, 20):
#     model = KMeans(n_clusters = i)
#     model.fit(x)
#     temp_list.append(model.inertia_)
#
# # 绘制图像
# plt.figure()
# plt.grid()
# plt.plot(range(1, 20),temp_list,'or-')
# plt.show()

# K = 5时效果最好

# 实例化模型
model = KMeans(n_clusters=5)
model.fit(x)
y_pred = model.predict(x)
print(y_pred)
print(model.cluster_centers_)

plt.figure()
plt.scatter(x.values[y_pred == 0, 0], x.values[y_pred == 0, 1],c = 'r',label = '1')
plt.scatter(x.values[y_pred == 1, 0], x.values[y_pred == 1, 1],c = 'b',label = '2')
plt.scatter(x.values[y_pred == 2, 0], x.values[y_pred == 2, 1],c = 'g',label = '3')
plt.scatter(x.values[y_pred == 3, 0], x.values[y_pred == 3, 1],c = 'gray',label = '4')
plt.scatter(x.values[y_pred == 4, 0], x.values[y_pred == 4, 1],c = 'y',label = '5')
plt.scatter(model.cluster_centers_[:, 0],model.cluster_centers_[:, 1], c = 'black', label = 'center')
plt.show()
```
